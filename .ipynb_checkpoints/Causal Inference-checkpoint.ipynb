<b> Causal Inference </b> 


The datset used for this assignment was same as the one I used in Assignment 1. The dataset contains information about a  telecom company's churn and was obtained from Kaggle ( https://www.kaggle.com/blastchar/telco-customer-churn). The dataset includes information about:
​
1. Customers who churned/ stopped using the company's services 
2. Services available to customers — phone, multiple lines, internet, online security, online backup, device protection, tech support, and streaming TV and movies
3. Customer account information — how long they’ve been a customer, the type of contract (month-to-month, one-year, two-years), payment method, paperless billing, monthly charges, and total charges
4. Demographic information of customers — gender, age, and if they have partners and dependents
​




Causal inference was conducted to test different hypothesis and detect if there were any causal relationships between different treatment options and the outcome (if a customer churns or not). The results from these different hypothesis will help in understanding if there are particular factors/features that are causing people to stop using the company's services/leaving the company. Understanding of these factors can help the company come up with a strategy to retain the consumers. These hypothesis were developed on the basis of the insights that were derived from the data exploration; for example, if a certain feature showed more or less correlation to churn. The feature selection process was also used to come up with hypothesis to see if the most important features were responsible for causation. Interface 2 steps from DoWhy documentation were followed to test different hypothesis. Please refer to the very end for the insights derived from the analysis.

# Importing Libraries

```python
import pandas as pd
import dowhy
from dowhy import CausalModel
from dowhy import plotter
```

# Loading the dataset

```python
df_copy = pd.read_csv(r"C:\Users\shiva\Downloads\archive (1)\WA_Fn-UseC_-Telco-Customer-Churn.csv")
df_copy.head()
```

# Feature Engineering

```python
#getting rid of missing values
df_copy = df_copy.dropna()

#Obtaining binary columns
binary_feature = df_copy.nunique()[df_copy.nunique() == 2].keys().tolist()
binary_feature.remove('SeniorCitizen') #Already has 1 and 0 
for i in binary_feature:
    df_copy[i] = df_copy[i].apply(lambda x: 1 if x=='Yes' else 0)
```

** Some of the variables although had three levels can be just subsetted such that they have two levels. For example, multiple lines had three levels: Yes, No, No Phone Service. In this case that could be paired up with people either having multiple lines or not considering no phone service still implies that people do not have multiple lines. Additionally, there is a separate variable called phone lines that takes care of this as well so no important inforrmation is left. Similarly, there are other variables that have same characterstics, hence all of these will be encoded in such a way that they can be either yes or no. 

```python
#getting rid of irrelevant features 
df_copy = df_copy.drop(columns='customerID')

##pairing together variables under yes and no 
df_copy['MultipleLines'] = df_copy['MultipleLines'].apply(lambda x: 1 if x=='Yes' else 0)
df_copy['OnlineSecurity'] = df_copy['OnlineSecurity'].apply(lambda x: 1 if x=='Yes' else 0)
df_copy['OnlineBackup'] = df_copy['OnlineBackup'].apply(lambda x: 1 if x=='Yes' else 0)
df_copy['DeviceProtection'] = df_copy['DeviceProtection'].apply(lambda x: 1 if x=='Yes' else 0)
df_copy['TechSupport'] = df_copy['TechSupport'].apply(lambda x: 1 if x=='Yes' else 0)
df_copy['StreamingTV'] = df_copy['StreamingTV'].apply(lambda x: 1 if x=='Yes' else 0)
df_copy['StreamingMovies'] = df_copy['StreamingMovies'].apply(lambda x: 1 if x=='Yes' else 0)

#creating dummy variables for all categorical variables 
df_copy = pd.get_dummies(df_copy, columns = [ 'InternetService', 'Contract', 'PaymentMethod'])
df_copy.head()

## Clubbing together all the services to reduce the number of common causes
df_copy['num_services'] = df_copy.OnlineBackup + df_copy.OnlineSecurity + df_copy.DeviceProtection +  df_copy.TechSupport + df_copy.StreamingMovies + df_copy.StreamingTV
```

```python
df_copy.columns
```

### Setting output column type to bool


```python
df_copy['Churn'] = df_copy['Churn'].apply(bool)
```

# Causal Inference: testing different hypothesis

## Hypothesis 1:  A person who has dependents can have an impact on his/her churning (stop using company's services)

 Assumption here is that the common causes, features that have causal relationship between both treatment and outcome is all features excluding  paperless billing and payment methods.

### Model

```python
df_copy['Dependents'] = df_copy['Dependents'].apply(bool)
```

```python
model= CausalModel(
        data=df_copy,
        treatment='Dependents',
        outcome="Churn",
        common_causes=['gender', 'SeniorCitizen', 'Partner', 'tenure',
       'PhoneService', 'MultipleLines', 'num_services',
        'MonthlyCharges', 'TotalCharges',  'InternetService_DSL', 'InternetService_Fiber optic',
       'InternetService_No', 'Contract_Month-to-month', 'Contract_One year', 'Contract_Two year'])
model.view_model(layout="dot")
```

### Identify

```python
identified_estimand = model.identify_effect(proceed_when_unidentifiable=True)
print(identified_estimand)
```

### Estimate

```python
causal_estimate = model.estimate_effect(identified_estimand,method_name="backdoor.linear_regression", test_significance = True)
print(causal_estimate)
print("Causal Estimate is " + str(causal_estimate.value))
```

### Refute

```python
refute_results=model.refute_estimate(identified_estimand, causal_estimate,
        method_name="random_common_cause")
print(refute_results)
```

# Conclusion

The main objective of running causal inference is to test a hypothesis. For this specific dataset, even though different hypothesis were conducted; they all were related to determining if a certain factor will affect churn or not. For this dataset, different features like tenure, having a dependent, having a partner, being a senior citizen, having a monthly contract etc. wete treated as treatment and the outcome was churn while the common causes included the features that could impact both treatment and outcome. Overall, the analysis was able to answer the problem statement or test the hypothesis in the sense that from the results, one can infer which feature would cause a customer to churn.

## Results

```python
results = {'Variables': ['Dependents', 'Phone Service', 'Partner', 'Internet Service - Fibre Optik', 'Contract - Month-to-Month','Tenure', 'Payment Method - Electronic Check','Gender', 'Paperless Billing', 'Senior Citizen'],
           'Causal Estimate': [ -0.009, 0.205, -0.091, 0.218, 0.030, -0.021, '-2.277*e-11', 0.001, 0.129, 0.098],
            'p value':[0.847, 0.044, 0.063, 0.022,0.656, 0.674, 0.449, 0.781, 0.001,0.108] }
results_df = pd.DataFrame(data=results)
results_df 
```

## Insights and Business Implications

The p value of less than 0.05 is considered statistically significant. As can be seen in the table above, the features with p value of less than 0.95 include:
<br> -Phone Service (if people had phone services), 
<br> -Internet Service - Fibre Optik (if customer's service provider is Fibre Optik), and
<br> -Paperless Billing (if the customer has paperless billing)

<br> Looking at the causal estimates of these features, it can be inferred that: 
<br> - The causal estimate for the phone service is 0.205, which implies that there is 20% more probabability of customers who have phone services to leave the company as compared to the ones who do not. 
<br> -  The causal estimate for the fibre optik internet service provider is 0.218, which implies that there is approximately 22% more probabability of customers who have fibre optik as their internet service provider to leave the company as compared to the ones who do not.
<br> - The causal estimate for paperless billing is 0.129, which implies that the likelihood of customers to stop using the company's services or churning out is almost 13% more if they have opted for paperless billing. 

*Note that for each of the models, refute step was performed as well to test the strength and validity of the causal effect found by the estimate step. Since the refute step gave almost identical values for the causal estimate as was generated by the estimate step, it is safe to say that the the company can base their strategy on these results. 

<br> <b> Business Insights: </b>
From the above results, we can say that there is causal relationship between people churning and people with phone services, people with fibre optik as their internet service provider and people who have paperless billing. The telecom company can use causal inference results to their advantage because now they can see where the real problem lies. People are unsatisfied with their phone services and fibre optik internet provider; hence the company needs to come up with a strategy to make those services better to retain the consumers. The parless billing being one of the factors that impact churn is a very interesting insight since opting for paperless billing is customer's decision and it will be hard for the company to push people to not go for paperless billing. However, the company can regularly contact consumers to check if they want to switch to an alternative way of billing. In the end, it can be seen that causal inference can really help with understanding the 'why' of the problem. In the analysis that was done for assignment 1, classification model was built, which could only predict if a customer will churn or not, but it didn't answer the question about why the customers who were predicted to churn would leave. Hence, the company need to test more hypothesis and perform causal inference to understand what is not working for consumers and try to fix that to drive customer retention.
